```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
coffee_dataframe = pd.read_csv("index_1.csv")
```

```python
coffee_dataframe
```

Scatter Plot

```python
import numpy as np
plt.grid(True)

plt.scatter(coffee_dataframe["money"], coffee_dataframe['coffee_name'], color="red")
plt.title("Coffee Price Data")
plt.xlabel("money")
plt.ylabel("coffee_name")
plt.xticks(np.arange(0, max(coffee_dataframe["money"]) + 1, 5))  

# save chart
plt.savefig("charts/coffee_price_data.png", dpi=300)

#show chart
plt.show()
```

Histogram

```python
plt.figure(figsize=(10, 6)) 

sns.histplot(data=coffee_dataframe, x="coffee_name")
plt.xticks(rotation=45, ha="right") 
sns.histplot(data=coffee_dataframe, x="coffee_name")
plt.title("Coffee Type Sale")
plt.xlabel("coffee_name")
plt.ylabel("count")

# save chart
plt.savefig("charts/coffee_type_sale.png", dpi=300)

#show chart
plt.show()
 
```

Kernel Density Plot

```python

plt.figure(figsize=(8, 6))
sns.kdeplot(data=coffee_dataframe, x="money",fill=True,  hue="coffee_name",common_norm=False, palette="crest", alpha=0.2)

plt.title("Kernel Density Plot of money by coffee name")
plt.xlabel("money")
plt.ylabel("Density")

# save chart
plt.savefig("charts/kernel_density_plot.png", dpi=300)

#show chart
plt.show()
```

Violin Plot

```python
sns.violinplot(x='cash_type', y='coffee_name', data=coffee_dataframe)
plt.title('Violin Plot of cash_type by coffee_name')

# save chart
plt.savefig("charts/violin_plot.png", dpi=300)

#show chart
plt.show()
```
